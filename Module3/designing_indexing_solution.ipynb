### Question 8

```python
### GRADED

### YOUR SOLUTION HERE
ans8a = None
ans8b = None

### BEGIN SOLUTION
ans8a = pd.merge(ans6a, ans7a, on = 'Country Code', how = 'inner')

ans8b = ans8a.isna().sum().sum()
ans8b
### END SOLUTION
```
