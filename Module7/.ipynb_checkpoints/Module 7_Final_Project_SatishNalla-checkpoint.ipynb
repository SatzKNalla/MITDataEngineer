### 2.3 Data Preparation

Determine if there are any Excess Nulls in the Data.


```python
#Figuting out the excess null variables from the data
ExcessNull = {}

for id,val in enumerate(data.columns):
    if data[val].isna().sum() > 0:
        #print(id,val, data[val].isna().sum())
        ExcessNull[val]=data[val].isna().sum()

ExcessNull = dict(sorted(ExcessNull.items(), key=lambda x: x[1], reverse=True))
ExcessNull
```

#### Dropping the Excess Null Value columns

Deleting the columns which has Null values for more than 30% of the observations, as we have 100 rows, any columns which has Nulls more than 30 are being deleted.

```python
#dropping out the excess null variables from the data
dropColumnList = []
totalObservations = data.shape[0]

for key in ExcessNull.keys():
    if ExcessNull[key]/totalObservations >= 0.3:
        print(f'Deleting the variable: {key} as it has {ExcessNull[key]} observations NULL')
        data.drop([key],axis=1,inplace=True)
```

#### Deleting the Unwanted columns

the columns like Unnamed, Id doesn't have any signifcance, so they are being deleted.

```python
#deleting the unwanted columns which are not significant
data.drop(['Unnamed: 0'],axis=1,inplace=True,errors='ignore')
data.drop(['Id'],axis=1,inplace=True,errors='ignore')
data.shape
```

As House Prices vary from size of the house, calculating the ***Price per sft***, by dividing Sale Price with Sum of 1stFloor SF and 2ndFloor SF Variables

```python
#creating all the additional possible metrics can be used to predict
data['SalePricePerSft'] = data['SalePrice']/(data['GrLivArea'])
data['LogSalePrice'] = np.log(data['SalePrice'])
data['LogSalePricePerSft'] = np.log(data['SalePricePerSft'])

MetricsList = ['SalePrice', 'SalePricePerSft','LogSalePrice','LogSalePricePerSft']

data.head()
```

### Converting the Text Variables to Number variables using from sklearn module

from sklearn by importing ***TransformerMixin*** class creating an conversion model from text ot numeric values

```python
#building the text variables to Number variables using sklearn module for our data
from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):

    def __init__(self):
        """Columns of dtype object are imputed with the most frequent value in column. Columns of other types are imputed with mean of column."""
    def fit(self, X, y=None):
        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in 
            X],index=X.columns)
        return self
    def transform(self, X, y=None):
        return X.fillna(self.fill)
X = pd.DataFrame(data) 
data = DataFrameImputer().fit_transform(X)
data.isna().sum()
```

```python
#iniating the data convesion from text to number
from sklearn import preprocessing
LE = preprocessing.LabelEncoder()
CategoryList = data.select_dtypes(include="object").columns
CategoryList
```

```python
#Applying the text to number model to our data 
for column in CategoryList:
    data[column] = LE.fit_transform(data[column])
data.head()
```

### 2.4 Correlation

The correlation coefficient is a statistical measure of the strength of a linear relationship between two variables. Its values can range from -1 to 1. A correlation coefficient of -1 describes a perfect negative, or inverse, correlation, with values in one series rising as those in the other decline, and vice versa. A coefficient of 1 shows a perfect positive correlation, or a direct relationship. A correlation coefficient of 0 means there is no linear relationship.
Correlation coefficients are used in science and in finance to assess the degree of association between two variables, factors, or data sets. For example, since high oil prices are favorable for crude producers, one might assume the correlation between oil prices and forward returns on oil stocks is strongly positive. Calculating the correlation coefficient for these variables based on market data reveals a moderate and inconsistent correlation over lengthy periods.

### 3.1 Linear Regression

Linear regression is a basic and commonly used type of predictive analysis.  The overall idea of regression is to examine two things: (1) does a set of predictor variables do a good job in predicting an outcome (dependent) variable?  (2) Which variables in particular are significant predictors of the outcome variable, and in what way do they–indicated by the magnitude and sign of the beta estimates–impact the outcome variable?  These regression estimates are used to explain the relationship between one dependent variable and one or more independent variables.  The simplest form of the regression equation with one dependent and one independent variable is defined by the formula y = c + b*x, where y = estimated dependent variable score, c = constant, b = regression coefficient, and x = score on the independent variable.

Naming the Variables.  There are many names for a regression’s dependent variable.  It may be called an outcome variable, criterion variable, endogenous variable, or regressand.  The independent variables can be called exogenous variables, predictor variables, or regressors.

Three major uses for regression analysis are (1) determining the strength of predictors, (2) forecasting an effect, and (3) trend forecasting.

As calculated the Correlation Co-efficient top 20 variables, we are using that as input for sklearn LinearRegression Algorithm to build the model
