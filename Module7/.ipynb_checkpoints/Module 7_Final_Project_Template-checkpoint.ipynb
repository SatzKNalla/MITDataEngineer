### 2.3 Data Preparation

Determine if there are any missing values in the data. Did the data need to be reshaped? If yes, include a description of the steps you followed to clean the data.


```python
import re
nulls = pd.DataFrame(data.isnull().sum().sort_values(ascending=False))
print(data.shape)

stringCols = data.select_dtypes(exclude=[np.number])
print(stringCols.shape)
print(stringCols.columns)

numberDict = {}
stringCols = list(stringCols)
stringCols = list(set([re.sub('_Numeric', '', fname) for fname in stringCols]))

for col in stringCols:
    temp = list(data[col])
    temp = set(temp)
    numberDict[col] = {}
    for id, val in enumerate(temp):
        numberDict[col][val]=id
    data[col+'_Numeric'] = data[col]
    data[col+'_Numeric'].replace(to_replace=numberDict[col], inplace = True)

data.shape
```

```python
data = data.select_dtypes(include=[np.number]).interpolate().dropna(axis = 1)
data.shape
```

```python
train = data.iloc[0:100,:]
print(train.shape)
train
```

```python
salePrice = train['SalePrice']
salePrice.describe()
plt.hist(salePrice)
```

```python
plt.hist(np.log(salePrice))
print(salePrice.skew())
print(np.log(salePrice).skew())
```

```python
target = np.log(salePrice)
plt.scatter(train['GrLivArea'],y=target)
```

```python
#plt.scatter(train['LotArea'],y=target)
train.head()

numeric = train.select_dtypes(include=[np.number])
print(len(numeric.columns))

i,j=0,0
PLOTS_PER_ROW = 5
fig, axs = plt.subplots(math.ceil(len(numeric.columns)/PLOTS_PER_ROW),PLOTS_PER_ROW, figsize=(20, 60))
for col in numeric.columns:
    axs[i][j].scatter(train[col],y=target, s=3)
    axs[i][j].set_ylabel(col)
    j+=1
    if j%PLOTS_PER_ROW==0:
        i+=1
        j=0

plt.show()

```

```python
#numeric = data.select_dtypes(include=[np.number])
numeric  = data
print(numeric.shape)
corr = numeric.corr()
cols = corr['SalePrice'].sort_values(ascending=False)
print(dict(cols))
cols = cols[0:10].index
print(cols)
```

```python
#X = train[cols]
X = data[cols]
Y = X['SalePrice']
X = X.drop(['SalePrice'],axis = 1)
X
```

```python
lr = linear_model.LinearRegression()
model = lr.fit(X, Y)
predictions = model.predict(X)
print(f'R^2 is: {model.score(X,Y)}')
```
