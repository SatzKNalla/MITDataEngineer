```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
list = np.random.rand(20)
ax.plot(list)
ax.set_title('First Graph')
ax.legend(list, loc="upper left")
plt.show()

n_rows = 2
n_cols = 2
fix, axes = plt.subplots(n_rows,n_cols)
for row_num in range(n_rows):
    for col_num in range(n_cols):
        ax = axes[row_num][col_num]
        list = []
        list = np.random.rand(20)
        ax.plot(list, color="blue")
        ax.legend(['Random Var'], loc="upper left")
        ax.set_title(f"Plot {row_num+1},{col_num+1}")

fix.suptitle("Main title")
fix.tight_layout()

plt.show()

```

```python
import matplotlib.pyplot as plt

courses = ['Data Science','Python','R','Statistics','Machine Learning']
students = [30, 45, 8, 25, 38]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.bar(courses, students)
plt.xlabel("Courses")
plt.ylabel("Student Enrollemnt")
plt.title("Student Enrollment by Course")
```

```python
import matplotlib.pyplot as plt
import numpy as np
N = 500
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x,y,c = 'g',alpha = 1)
plt.title('Scatter plot using Matplotlib')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
